```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_data.csv')
```

```python
label_encoder = LabelEncoder()
df['Loại Nhà (mã hóa)'] = label_encoder.fit_transform(df['Loại nhà'])
encoded_values_loai = df[['Loại nhà', 'Loại Nhà (mã hóa)']].drop_duplicates().sort_values('Loại Nhà (mã hóa)')
print(encoded_values_loai)
```

```python
df['Quận (mã hóa)'] = label_encoder.fit_transform(df['Quận'])
encoded_values_quan = df[['Quận', 'Quận (mã hóa)']].drop_duplicates().sort_values('Quận (mã hóa)')
print(encoded_values_quan)
```

```python
df['Khoảng giá'] = df['Mức giá (tỷ)'].apply(lambda x: 0 if x <= 2 else (1 if x < 10 else (2 if x < 50 else 3)))

df_low = df[df['Khoảng giá']==0]
df_low_avg = df[df['Khoảng giá']==1]
df_high_avg = df[df['Khoảng giá']==2]
df_high = df[df['Khoảng giá']==3]
print(df_low.shape[0])
print(df_low_avg.shape[0])
print(df_high_avg.shape[0])
print(df_high.shape[0])
```

```python
# Bổ sung thêm cột Mức giá (triệu/m²)
df['Mức giá (triệu/m²)'] = df['Mức giá (tỷ)'] * 1000 / df['Diện tích (m²)']

# Xóa tất cả các dòng có giá trị quá lớn trong cột 'Mức giá (triệu/m²)' (lớn hơn 2000)
df = df.drop(df[df['Mức giá (triệu/m²)'] > 2000].index)
```
